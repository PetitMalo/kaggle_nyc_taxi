## Setting paths

```python
notebook_dir = Path(os.getcwd())
root_dir = notebook_dir.parent if notebook_dir.name == "src" else notebook_dir
data_dir = root_dir / "data"

train_path = data_dir / "train.csv"
test_path = data_dir / "test.csv"

print(f"Notebook dir: {notebook_dir}")
print(f"Root dir: {root_dir}")
print(f"Data dir: {data_dir}")
print(f"Train path: {train_path}")
print(f"Train exists: {train_path.exists()}")
```

Converting CSVs to parquets

```python
if not train_path.with_suffix(".parquet").exists():
    df_train = pd.read_csv(train_path)
    df_train.to_parquet(data_dir / "train.parquet")

if not test_path.with_suffix(".parquet").exists():
    df_test = pd.read_csv(test_path)
    df_test.to_parquet(data_dir / "test.parquet")
```
